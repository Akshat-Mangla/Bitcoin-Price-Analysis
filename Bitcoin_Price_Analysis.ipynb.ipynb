```python
import matplotlib.pyplot as plt
from pycoingecko import CoinGeckoAPI

cg = CoinGeckoAPI()

# Fetch Bitcoin market chart data for the last 30 days
data = cg.get_coin_market_chart_by_id(id='bitcoin', vs_currency='usd', days=30)

# Extract prices and timestamps
timestamps = [point[0] for point in data['prices']]
prices = [point[1] for point in data['prices']]

# Convert timestamps to readable dates
import datetime
dates = [datetime.datetime.fromtimestamp(ts/1000).strftime('%Y-%m-%d') for ts in timestamps]

# Plot the data
plt.figure(figsize=(10, 5))
plt.plot(dates, prices, label='Bitcoin Price (USD)', color='blue')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.title('Bitcoin Price Over Last 30 Days')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
